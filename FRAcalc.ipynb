```python
from myABBR import * ; import myUtil as mu
tradeDT   = jDT(2025,12,19) ; setEvDT(tradeDT)
tbCrvDATA = [('depo','6M',1.07076), ('fra','1M',1.115), ('swap','2Y',1.43)]
tbrIX, tbCrvOBJ, tbCrvHDL, tbParRT = mu.makeTiborCurve(tbCrvDATA)
# checking
print('tradeDT:{}, settDT:{}'
                .format(tradeDT.ISO(),tbCrvOBJ.referenceDate().ISO()))
dfTBR = pd.DataFrame(tbCrvOBJ.nodes(), columns=['node','DF'])
display(dfTBR[:8].style.format(fmtSCF))
```

* 6mTibor

```python
# 6mTibor     (DTxxは日付xxを表す変数)
DT2d = calJP.advance   (tradeDT, 2, DD)         # スタート日 
DT6m = calJP.advance   (DT2d,    6, MM)         # 満期日
tbTNR = dcA365.dayCount(DT2d,     DT6m)         # テナー日数
DF6m  = 1 / (1+tbParRT[0]*tbTNR/365)

print(f'tb6m:{tbParRT[0]:.6%},   ' f'DF6m:{DF6m:.8f}\n'       
      f'(スタート日:{DT2d.ISO()}, ' f'テナー日数:{tbTNR}日 )' )
```

* 1 x 7 FRA

```python
# 1x7 FRA
DT1m  = calJP.advance  (DT2d, 1, MM)                  # スタート日
DT7m  = calJP.advance  (DT1m, 6, MM)                  # 満期日
frTNR = dcA365.dayCount(DT1m, DT7m)                   # テナー日数
DF1m  = tbCrvOBJ.discount(DT1m)                       # スタート日DF
DF7m  = DF1m / (1+tbParRT[1]*frTNR/365)

print(f'fra1x7:{tbParRT[1]:.6%}, '  f'DF1m:{DF1m:.8f}, '
      f'DF7m:{DF7m:.8f}\n'       
      f'(スタート日:{DT1m.ISO()}, '  f'テナー日数:{frTNR}日 )' )
```
