```python
from myABBR import * ; import myUtil as mu
# 0. 初期値設定        (JSDAは当日決済の計算の為、tradeDTを1日前に設定)
tradeDT,         mPRC,     mYLD,         effDT,         matDT,     cpnRT  =\
jDT(2025,11,17), 103.37, 0.925/100, jDT(2008,5,30), jDT(2028,3,20), 2.4/100

# 1. スケジュールオブジェクトと債券オブジェクト
bndSCD = ql.Schedule(effDT, matDT, pdFreqSA, calJP, unADJ,unADJ, dtGENb, EoMf)
JL101  = mu.JGB(Tp1, parAMT, bndSCD, [cpnRT], dcA365n)
JL101a = mu.JGB(Tp1, parAMT, bndSCD, [cpnRT], dcA365 )

# 2. 各種日数
settDT   = calJP.advance      (tradeDT, Tp1, DD) ; setEvDT(tradeDT)
preCpnDT = bndSCD.previousDate(settDT)
accDS    = dcA365.dayCount    (preCpnDT, settDT)

# 3. 利回り計算等
sPRC   = JL101.SimplePrice       (mYLD)
sYLD   = JL101.SimpleYield       (mPRC)
cYLD   = JL101.JapanCompoundYield(mPRC)
rYLD   = JL101.bondYield      (cP(mPRC), dcA365n, cmpdCMP, freqSA)
accAMT = JL101a.accruedAmount    (    )

print('                         ++ No rounddown ++'   )
print('{:<9} : {:>8.2f} ; {:>12.7f}   at {}'.format(
'単利->価格',  rD(sPRC,2),   sPRC,     settDT.ISO()   ))
print('{:<8} : {:>8.3%} ; {:>12.7%}'.format(
'価格->日本複利', rD(cYLD,5), cYLD                     ))
print('{:<8} : {:>5}    ; {:>12.7%}'.format(
'価格->欧米複利', '',         rYLD                     ))
print('{:<10}: {:>8.3%} ; {:>12.7%}'.format(
'価格->単利',  rD(sYLD,5),   sYLD                      ))
print('{:<10}: {:>5}    ; {:>12.8f}   at {}日'.format(
'経過利息',      '',       accAMT,       accDS         ))
```
